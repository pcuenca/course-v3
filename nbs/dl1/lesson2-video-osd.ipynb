```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
def predict_frame(frame, learn):    
    # Convert frame to Tensor
    swapped = np.swapaxes(frame, -1, 0)
    swapped = np.float32(swapped / 255.)
    pixel_tensor = torch.from_numpy(swapped)
    
    img = Image(pixel_tensor)
    pred_class, pred_idx, outputs = learn.predict(img)
    #print(type(pred_class.obj))
    #return pred_class.obj
    return outputs

output = predict_frame(frame, learn)
label = 'Darth Vader: {:.3}    Luke: {:.3}'.format(output[0], output[1])
label
```

```python
video_in = 'data/father.avi'
video_out = 'data/output.avi'

# Based on snippet by David Griñán
cap = cv2.VideoCapture(video_in)

round(cap.get(cv2.CAP_PROP_FPS))
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter(video_out, fourcc, cap.get(cv2.CAP_PROP_FPS), (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        #label = predict_frame(frame, learn)
        output = predict_frame(frame, learn)
        
        label = 'Darth Vader: {:.3}    Luke: {:.3}'.format(output[0], output[1])
        print(label)
        
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, label, (10,50), font, 2, (255,255,255), 2, cv2.LINE_AA)

        out.write(frame)
    else:
        print("ko")
        break
```
